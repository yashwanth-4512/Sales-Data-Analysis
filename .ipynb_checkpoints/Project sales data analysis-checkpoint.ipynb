```python
import pandas as pd
import numpy as np
import io
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
```

```python
# Suppress warnings for cleaner output
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
```

```python
# --- Configuration ---
RANDOM_SEED = 42
DATE_COLUMN = 'date'
TARGET_COLUMN = 'sales'
AGGREGATED_TARGET = 'Daily_Sales'
TEST_SIZE_RATIO = 0.2
MINIMAL_TRAIN_DAYS = 60 # Minimum days required for lag/rolling features
```

```python
print("Phase 1: Data Preparation and Feature Engineering...")

# 1. Data Ingestion and Initial Exploration

# Load the full data from the fetched content.
df_full = pd.read_csv("C:\\Users\\yashw\\PyCharmMiscProject\\Sales Data Analysis\\demand-forecasting-kernels-only\\train.csv", parse_dates=[DATE_COLUMN])

# --- Data Cleaning and Aggregation ---

# Check for missing values (No missing values found in this dataset after inspection)
# assert df_full.isnull().sum().sum() == 0

# The target is 'daily total sales' across all stores and items.
df_daily = df_full.groupby(DATE_COLUMN)[TARGET_COLUMN].sum().reset_index()
df_daily.rename(columns={TARGET_COLUMN: AGGREGATED_TARGET}, inplace=True)
df_daily.set_index(DATE_COLUMN, inplace=True)
df = df_daily.sort_index()

print(f"Data aggregated to Daily Sales. Total days: {df.shape[0]}")
```

```python
# 2. Feature Engineering
# Temporal Features
print("Creating Temporal Features...")
df['Year'] = df.index.year
df['Month'] = df.index.month
df['DayOfWeek'] = df.index.dayofweek
df['DayOfYear'] = df.index.dayofyear
df['Day'] = df.index.day
df['WeekOfYear'] = df.index.isocalendar().week.astype(int)

# Holiday Flag (Simplified proxy: High sales volume days across all historical data)
# This uses future data, which is an acceptable proxy if a real holiday calendar isn't available.
# A more robust solution would require external holiday data.
overall_sales_threshold = df[AGGREGATED_TARGET].quantile(0.99)
df['Is_High_Sales_Day'] = (df[AGGREGATED_TARGET].shift(periods=1) > overall_sales_threshold).astype(int)
df['Is_High_Sales_Day'] = df['Is_High_Sales_Day'].fillna(0) # Fill first NaN
```

```python
# Lagged Features
print("Creating Lagged Features...")
for lag in [1, 7, 30, 90]:
    df[f'Sales_Lag_{lag}'] = df[AGGREGATED_TARGET].shift(lag)

# Rolling Window Features (using mean and std deviation to capture trend and volatility)
print("Creating Rolling Window Features...")
for window in [7, 30]:
    # Shift by 1 period before rolling window calculation to avoid data leakage
    df[f'Sales_RollMean_{window}'] = df[AGGREGATED_TARGET].shift(1).rolling(window=window).mean()
    df[f'Sales_RollStd_{window}'] = df[AGGREGATED_TARGET].shift(1).rolling(window=window).std()

# Drop rows with NaN values introduced by lagging/rolling (typically the first 90 days)
df.dropna(inplace=True)
print(f"Feature matrix successfully created. Final data shape: {df.shape}")

# Define X (Features) and y (Target)
X = df.drop(columns=[AGGREGATED_TARGET])
y = df[AGGREGATED_TARGET]
```

```python
# --- Phase 2: Model Development and Training (Random Forest) ---
print("\n" + "="*70)
print("Phase 2: Model Development and Training...")

# 1. Data Splitting (Time-series split: Train on older, Test on newer)
split_point = int(len(df) * (1 - TEST_SIZE_RATIO))
X_train = X.iloc[:split_point]
X_test = X.iloc[split_point:]
y_train = y.iloc[:split_point]
y_test = y.iloc[split_point:]

print(f"Training period: {X_train.index.min().date()} to {X_train.index.max().date()}")
print(f"Testing period: {X_test.index.min().date()} to {X_test.index.max().date()}")
```

```python
# 2. Model Selection and Hyperparameter Tuning (using GridSearchCV)
rf_model = RandomForestRegressor(random_state=RANDOM_SEED)

# Define parameter grid: smaller grid for efficiency but covering key parameters
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 15],
    'min_samples_leaf': [1, 5]
}

print("Starting Hyperparameter Tuning (GridSearch)...")
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid,
                           cv=3, scoring='r2', n_jobs=-1, verbose=0)
grid_search.fit(X_train, y_train)

best_rf_model = grid_search.best_estimator_
print(f"Best Random Forest Parameters: {grid_search.best_params_}")
print("Phase 2 Complete. Model Trained.")
```

```python
# --- Phase 3: Evaluation and Reporting ---
print("\n" + "="*70)
print("Phase 3: Evaluation and Reporting...")

# 1. Prediction and Performance Evaluation (Random Forest)
y_pred_rf = best_rf_model.predict(X_test)
final_r2_rf = r2_score(y_test, y_pred_rf)
final_mae_rf = mean_absolute_error(y_test, y_pred_rf)

print(f"\n--- Random Forest Model Performance ---")
print(f"ðŸŽ¯ Final RÂ² Score (Coefficient of Determination): **{final_r2_rf:.4f}**")
print(f"Mean Absolute Error (MAE): {final_mae_rf:.2f}")
```

```python
# 2. Benchmark Comparison (Naive Forecast: Sales_Lag_1)
# The naive forecast simply assumes tomorrow's sales = yesterday's sales.
y_pred_naive = X_test['Sales_Lag_1'].values
r2_naive = r2_score(y_test, y_pred_naive)

# Calculate Improvement: (Model R2 - Benchmark R2) / Benchmark R2
improvement_ratio = (final_r2_rf - r2_naive) / r2_naive * 100

print(f"\n--- Benchmark Performance (Naive Forecast) ---")
print(f"Naive Forecast (Lag 1) RÂ² Score: {r2_naive:.4f}")
print(f"Percentage Improvement of RF over Naive: **{improvement_ratio:.2f}%**")
```

```python
# 3. Visual Reporting (Matplotlib)

# 3a. Actual Sales vs. Predicted Sales Plot
plt.figure(figsize=(14, 6))
plt.plot(y_test.index, y_test.values, label='Actual Sales', color='tab:blue', linewidth=2)
plt.plot(y_test.index, y_pred_rf, label='Predicted Sales (Random Forest)', color='tab:red', linestyle='--', linewidth=2)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.title('Actual vs. Predicted Sales (Test Set) ', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Daily Total Sales Volume')
plt.legend()
plt.grid(True, which='both', linestyle=':', alpha=0.6)
plt.show()

# 3b. Feature Importance Plot
importances = best_rf_model.feature_importances_
feature_names = X_train.columns
forest_importances = pd.Series(importances, index=feature_names).sort_values(ascending=False)

plt.figure(figsize=(12, 7))
# Plotting top 15 features for clarity
top_features = forest_importances.head(15)
top_features.plot.barh(color='tab:green')
plt.title("Top 15 Feature Importances (Random Forest) ", fontsize=16)
plt.xlabel("Feature Importance (Mean Decrease in Impurity)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()


print("\n" + "="*70)
print("FINAL DELIVERABLES")
print(f"Final RÂ² Score (Random Forest): {final_r2_rf:.4f}")
print(f"Improvement over Naive Benchmark: {improvement_ratio:.2f}%")
print("Complete Python Code: Provided above (including data loading and processing).")
print("Visual Reports: Two Matplotlib plots displayed.")
```
